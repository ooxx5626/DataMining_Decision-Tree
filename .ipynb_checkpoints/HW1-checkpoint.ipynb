```python
import pandas as pd # 引用套件並縮寫為 pd  

import os

# print(os.getcwd())
DIR=os.getcwd()
df = pd.read_csv(DIR+'\\train_big.csv')
# print(df.iloc[0:2,0:3])
```

```python
data = df.iloc[:,0:5]
target =  df.iloc[:,6]
feature_names =["Department and curriculum closeness rate","Attendance rate","HW score","Report score","number of ask teacher or TA"]
# feature_names =df.head(0)
target_names = ["A","B","C","D","E"] #,"F","G","H","I","J"
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(data, target)
```

```python
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=feature_names,  
                         class_names=target_names,  
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

```python
def visualize_classifier(model, X, y, ax=None, cmap='rainbow'):
    ax = ax or plt.gca()
    
    # Plot the training points
    ax.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, s=30, cmap=cmap,
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # fit the estimator
    model.fit(X, y)
    xx, yy = np.meshgrid(np.linspace(*xlim, num=200),
                         np.linspace(*ylim, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    # Create a color plot with the results
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.5,
                           levels=np.arange(n_classes + 2) - 0.5,
                           cmap=cmap,
                           zorder=1)

    ax.set(xlim=xlim, ylim=ylim)
from sklearn import tree
# clf = tree.DecisionTreeClassifier().fit(X, y)
# visualize_classifier(clf, df.iloc[:,[0,5]], y) #Department and curriculum closeness rate
```

```python
import numpy as np
import matplotlib.pyplot as plt

importances = clf.feature_importances_
# get sort indices in descending order
indices = np.argsort(importances)[::-1]
X_train = df.iloc[:,:5]
for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, 
                            df.columns.values[indices[f]], 
                            importances[indices[f]]))

plt.figure()
plt.title('Feature Importances')
plt.bar(range(df.iloc[:,:5].shape[1]),
        importances[indices],
        align='center',
        alpha=0.5,
        color='#FF8800')

plt.xticks(range(X_train.shape[1]), 
           X.columns.values[indices], rotation=100)
plt.xlim([-1, X_train.shape[1]])
plt.tight_layout()
plt.savefig('./output/fig-forest-feature-importances.png', dpi=300)
plt.show()
```
